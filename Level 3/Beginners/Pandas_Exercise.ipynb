Question 9:Create two data frames using the following two Dicts, Concatenate those two data frames and create a key for each data frame.

GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}

```python
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}
```

```python
GermanCars = pd.DataFrame(GermanCars)
GermanCars
```

```python
japaneseCars = pd.DataFrame(japaneseCars)
japaneseCars
```

```python
#concatenating
cars = pd.concat([GermanCars, japaneseCars], keys = ['GermanCars', 'japaneseCars'], names=['Key', 'ID'])
cars
```
