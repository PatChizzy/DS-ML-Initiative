```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('company_sales_data.csv')
data
```

Exercise 1: Read Total profit of all months and show it using a line plot

```python
x = data['month_number']
y = data['total_profit']
plt.plot(x,y)
plt.xlabel('Month [In Numbers]')
plt.ylabel('Total Profit')
plt.title('Total Profit by Month')
plt.show()
```

Exercise Question 2: Read all product sales data and show it  using a multi line plot

```python
month                  = data ['month_number']
face_cream             = data ['facecream']
face_wash              = data ['facewash']
tooth_paste            = data ['toothpaste']
bathing_soap           = data ['bathingsoap']
shampoo                = data ['shampoo']
moisturizer   = data ['moisturizer']
                 
plt.plot(month, face_cream,   label = 'Face cream Sales Data', marker='o', linewidth=3)
plt.plot(month, face_wash,   label = 'Face Wash Sales Data', marker='o', linewidth=3)
plt.plot(month, tooth_paste, label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(month, bathing_soap, label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(month, shampoo, label = 'ToothPaste Sales Data', marker='o', linewidth=3)
plt.plot(month, moisturizer, label = 'ToothPaste Sales Data', marker='o', linewidth=3)

                 
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.xticks(month)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Sales data')
plt.show()
```

Exercise Question 3: Read toothpaste sales data of each month and show it using a scatter plot

```python
x = data['month_number']
y = data['toothpaste']
plt.scatter(x,y)
plt.xlabel('Months')
plt.ylabel('Toothpaste')
plt.title('Toothpaste Sales by Month')
plt.xticks(month)
plt.grid(True, linewidth= 1, linestyle="--")
plt.show()
```

Exercise Question 5: Calculate total sale data for last year for each product and show it using a Pie chart

```python
elements      = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
Total_sales   = [data ['facecream'].sum(), data ['facewash'].sum(), data ['toothpaste'].sum(), 
                 data ['bathingsoap'].sum(), data ['shampoo'].sum(), data ['moisturizer'].sum()]

plt.axis("equal")
plt.pie(Total_sales, labels=elements, autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Sales data by Item')
plt.show()
```
